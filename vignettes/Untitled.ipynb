```python
data=pd.read_csv('../data/dataset 1.txt', sep='\t')
```

```python
plt.plot(data.time,data.X)
plt.show()
```

```python
data2=EWS.check_time_series(data.loc[:,'X'],data.time)
```

```python
data3=data2[0]
```

```python
result=EWS.EWS(data3,window_size=5000,autocorrelation=True,variance=True,skewness=True)
```

```python
print(EWS.kendalltrend(result.dropna().autocorrelation))
print(EWS.kendalltrend(result.dropna().variance))
print(EWS.kendalltrend(result.dropna().skewness))
```
